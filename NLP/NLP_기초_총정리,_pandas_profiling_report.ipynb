  ## 1-3) NLP에서 특별한 점은 무엇이 있는가?
  - Natural Language 즉 인간의 언어는 무언가를 표현하려고 한다. 즉 내가 말해서 어떤 정보를 상대에게 전달하는것이 기본적인 기능이다. 이 기능은 다른 관점에서 보면 일종의 [신호 시스템] 이라고 볼 수 있다.
  - Discrete / Symbolic / Categorical 다양한 관점에서 볼 수 있다. 즉 내가 전달하고자 하는 바를 최대한 정확하게 전달하는 신호체계 라는 뜻이다.
  - 예외적으로 특정 부분을 강조하기 위해 반어법을 사용한다거나, 비꼰다거나 하는 경우가 있다. 혹은 소리나 제스처, 글자를 상형화 혹은 신호화 하여 소통하는 경우도 있다. 이런 부분은 상당히 어려운 부분이다. 이들을 인코딩 즉 신호화 및 변환 하는 과정( 혹은 이들을 찾아내는 과정 )이 어렵고 복잡하다.
  - NL은 뇌에서 신호화하는 activation의 연속적인 패턴으로 볼 수 있다. 각자의 뇌가 다 다르지만, 그 기본적인 구조가 비슷하다면, 언어의 상징적/분류적 신호체계는 많은 단어, 이들의 신호화 , 이들의 기계학습 과정에서 sparsity 문제를 야기시킨다.

  ## 1-4) NLP의 단점 및 어려운 이유
    - 복잡하다. ( 표현 방법의 차이 , 세계별로 다른 문화와 가치관에 따른 다른 지식 및 정보, 언어는 진화하고, 변형되지만 이를 표준화하거나 사전화하기 힘든 부분 )
    - 애매하다. ( NL은 프로그래밍언어처럼 적당한 form 혹은 범위 안에서만 존재하는 것이 아니라, 지속적으로 변형되거나 혼합된다.)
    - 의존적이다. ( NL 특히 구어체쪽을 해석하기 위해서는 실제 상식 및 지식, 문맥정보, 대화의 상황 및 시간/장소 등을 알아야 정확하게 해석할 수 있다. )
    - 특히 직관적인 언어 분석의 단점 관련하여
      1. 단어의 세부적인 뉘앙스 차이를 놓치기 때문이다.
  예를 들어, "훌륭한"이라는 단어가 있을 때 good이라는 표현이 있지만 동의어(synonyms)이지만 그 의미하는 정도에서 차이가 있는 'adept', 'expert', 'proficient', 'skillful'과 같은 다른 단어들이 있다.

      2. 새로운 단어를 놓칠 수 있다. 항상 최신의 상태로 그 의미를 담고 있지는 않기 때문이다.
  예를 들면 'genius', 'wizard', 'ninja' 같이 새로운 단어는 '훌륭한'이라는 의미를 갖게 된지 얼마되지 않았다면 이를 반영하지 못할 수도 있다.

      3. 그리고 상당히 주관적인(subjective) 방법이기도 하다.

      4. 인간이 끊임없이 만들고 수정해야 하는 문제가 있다.

      5. 단어간의 유사성을 정확하게 계산하기가 어렵다.

###### 상관관계의 종류
- 1. Pearson (피어슨)
  - 상관분석에서 가장 기본적으로 사용되는 상관계수
  - 연속형 변수의 상관관계를 측정한다.( ex. 키, 몸무게 )
  - 모수 검정( parametric test ) - 고정된 모수의 집합에 대해 적절하게 표본 모델링 한다는 의미

- 2. Kendall (켄달)
  - 켄달 상관 계수 Tau
  - 변수값 대신 순위로 바꿔서 이용하는 상관계수(학교등급, 졸업학위 level)
  - 비모수 검정( non-parametric test)
  - 샘플사이즈가 적거나, 데이터의 동률이 많을 때 유용하다.

- 3. Spearman (스피어만)
  - 스피어만 상관 계수 rho
  - 변수값 대신 순위로 바꿔서 이용하는 상관계수(학교등급, 졸업학위 level)
  - 비모수 검정( non-parametric test)
  - 데이터의 편차와 에러에 민감하다. 일반적으로 켄달보다 높은 상관관계를 보인다.
  - 켄달과 스피어만은 순위 일치 정도를 검사한다는 점에서 유사하다.
