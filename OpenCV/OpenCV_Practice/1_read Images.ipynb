## Read Media FIle

```python
import cv2

video_file = 'Ryan_Cafe.mp4'

vcap = video_capture_obj = cv2.VideoCapture(video_file)
if vcap.isOpened():
    while True:
        ret, img = vcap.read()
        if ret:
            cv2.imshow(video_file, img)
            cv2.waitKey(25)# 25ms 지연 = 40fps , 25보다 작은 수를 넣으면 영상이 빠르게 재생되는 것처럼 보인다. 이를 지연시간이라고 한다.
            # FPS는 그래픽카드가 1초에 뽑아내는 화면의 수를 말한다. FPS가 높을수록 보여지는 이미지의 양이 더 많다.
        else:
            break
else:
    print('Can;t open Video File')

vcap.release()# 객체 자원 반남
cv2.destroyAllWindows()# 이 줄을 빼면 무한루프로 동영상 창이 꺼지지 않음( 정확히는 꺼도 계속 켜짐 )
```

## 만약 PC 및 노트북에 카메라 혹은 웹캠이 잇을때, 이를 연동한다면

```python
import cv2

cap = cv2.VideoCapture(0)# 0번 카메라 장치 연결 (디폴트값)
if cap.isOpened():
    while True:
        ret, img = cap.read()
        if ret:
            cv2.imshow('camera', img)
            if cv2.waitKey(1) != -1:
                break
        else:
            print('no frame')
            break
else:
    print("Can't open camera")
cap.release()
cv2.destroyAllWindows()
```

## 웹캠으로 특정 프레임의 이미지 저장하기(셀카찍기 같은 느낌)

```python
import cv2

cap = cv2.VideoCapture(0)# 0번 카메라 장치 연결 (디폴트값)
if cap.isOpened():
    while True:
        ret, frame = cap.read()
        if ret:
            cv2.imshow('camera', frame)
            if cv2.waitKey(1) != -1:
                cv2.imwrite('photo.jpg', frame)
                break
        else:
            print('no frame')
            break
else:
    print('No Camera')
cap.release()
cv2.destroyAllWindows()
```

## 웹캠으로 녹화하기

```python
import cv2

cap = cv2.VideoCapture(0)
if cap.isOpened:
    file_path = ''#저장할 파일 경로
    fps = 30.0#  FPS 초당 프레임 수
    fourcc = cv2.VideoWriter_fourcc(*'DIVX')
    width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
    height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
    size = (int(width), int(height))
    
    out = cv2.VideoWriter(file_path, fourcc, fps, size)
    while True:
        ret, frame = cap.read()
        if ret:
            cv2.imshow('camera-recording', frame)
            out.write(frame)
            if cv2.waitKey(int(1000/fps)) != -1:
                break
        else:
            print('No Frame')
            break
    out.release()
else:
    print("Can't open Camera")
cap.release()
cap.destroyAllWindows()
            
```
